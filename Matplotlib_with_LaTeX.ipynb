```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

# Normal Matplotlib Fonts

```python
x_plot    = np.arange(0, 10, 1) # time (s)
slope     = 3.0 # accerlation
intercept = 4.0 # initial speed in m/s
y_plot    = slope * x_plot + intercept # velocity 
```

```python
fontsize = 20
plt.scatter(x_plot, y_plot)
plt.xlabel('Time (s)', fontsize=fontsize)
plt.ylabel('Velocity (km/s)', fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.tight_layout()
plt.savefig('demo_normal_font.pdf')
plt.savefig('demo_normal_font.png')
plt.show()
```

# We can use LaTeX equation format in the label in Matplotlib environment and label a LaTeX equation

```python
fontsize = 20
label = r'$v(t) = a \times t + v_0$'
plt.scatter(x_plot, y_plot)
plt.text(0, 28, label, fontsize=fontsize)
plt.xlabel('Time (s)', fontsize=fontsize)
plt.ylabel('Velocity (km s$^{-1}$)', fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.tight_layout()
plt.savefig('demo_normal_font_mathmode.pdf')
plt.savefig('demo_normal_font_mathmode.png')
plt.show()
```

# Use LaTeX Font in Matplotlib

```python
plt.rcParams.update({"text.usetex": True})
```

```python
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
```

```python
fontsize = 20
label = r'$v(t) = a \times t + v_0$'
plt.scatter(x_plot, y_plot)
plt.text(0, 28, label, fontsize=fontsize)
plt.xlabel('Time (s)', fontsize=fontsize)
plt.ylabel('Velocity (km s$^{-1}$)', fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.tight_layout()
plt.savefig('demo_latex_font.pdf')
plt.savefig('demo_latex_font.png')
plt.show()
```
